```python
class BinaryConverter:
    @staticmethod
    def string_to_binary(s):
        return ' '.join(format(ord(c), '08b') for c in s)
    @staticmethod
    def binary_to_string(b):
        b = b.replace(" ", "")
        binary_int = int(b, 2)
        byte_number = (binary_int.bit_length() + 7) // 8
        binary_array = binary_int.to_bytes(byte_number, "big")
        return binary_array.decode()
```

```python
converter = BinaryConverter()
```

```python
binary_e = "01100110 01110010 01101111 01111010 01100101 01110010 00110000 00110100 01110110 01101001 01100011 01000000 01100111 01101101 01100001 01101001 01101100 00101110 01100011 01101111 01101101"
email= converter.binary_to_string(binary_e)
binary_pass="00111001 00111000 00110011 00110100 01000110 01110010 01101111 01111010 01100101 01110010 00110001 00110001 00110000 00110101"
passwd=converter.binary_to_string(binary_pass)
```

```python
import getpass
from hugchat import hugchat
from hugchat.login import Login
sign = Login(email, passwd)
cookies = sign.login()
cookie_path_dir = "./cookies_snapshot"
sign.saveCookiesToDir(cookie_path_dir)
despedida=["salir","adios","good bye","sayonara baby","hasta luego"]
```

```python
def start_conversation():
    chatbot = hugchat.ChatBot(cookies=cookies.get_dict())
    conversation_id = chatbot.new_conversation()
    print("ChatBot: Hola, soy un ChatBot. ¿En qué puedo ayudarte?")
    while True:
        user_message = input("Usuario: ")
        if user_message.lower() in despedida:
            print("ChatBot: Hasta luego.")
            break
        chatbot.change_conversation(conversation_id)
        response = chatbot.query(user_message)
        print("ChatBot:", response)
start_conversation()
```
